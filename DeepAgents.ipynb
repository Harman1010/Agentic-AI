```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from typing import Literal
from tavily import TavilyClient
tavily_client = TavilyClient(api_key=os.getenv("TAVILY_API_KEY"))
def web_search(query,max_results,topic:Literal["Sports","Economics","Finance"] = "general",include_raw_content=False):
    """Returns web search"""
    tavily_Search = tavily_client.search(query,max_results=5,topic=topic,include_raw_content=include_raw_content)
    return tavily_Search
    
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("google_genai:gemini-2.5-flash")
```

```python
from deepagents import create_deep_agent

agent = create_deep_agent(
    model = model,
    tools = [web_search],
    system_prompt = "You are an expert researcher. Use the web search tool to find information."
)
```

```python
response = agent.invoke({"messages" : [{"role":"user","content":"What is the recent news about Cricket sports?"}]})
print(response["messages"][-1].content)
```
