```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
import re

def determinstic_guardrail(input : str):
    """Returns True if found"""
    banned_keywords = ["hack","malware","bomb"]
    return any(kw in input.lower() for kw in banned_keywords)
text_inputs = [
        "What is AI?",
        "How to hack a server",
        "What do you understand by Plastic?"
    ]
for t in text_inputs:
        deterministic = determinstic_guardrail(t)
        print(f"Blocked: {t}") if deterministic else print(f"Safe: {t}")
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import SystemMessage, HumanMessage

model = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)

def modelBasedGuardrail(user_input: str):
    """Uses an LLM to safeguard the inputs"""
    # 1. SystemMessage sets the rules
    system_instruction = SystemMessage(
        content="You are an expert classifier. You have to classify the inputs as Safe or Unsafe."
    )
    # 2. HumanMessage passes the data
    user_message = HumanMessage(
        content=f"input: {user_input}"
    )
    
    # Pass both messages
    result = model.invoke([system_instruction, user_message])
    return result.content

# Example usage
text_inputs = ["How to bake a cake?", "How to make a bomb?"]
for t in text_inputs:
    result = modelBasedGuardrail(t)
    print(f"Input: {t} -> Guardrail: {result}")
```
